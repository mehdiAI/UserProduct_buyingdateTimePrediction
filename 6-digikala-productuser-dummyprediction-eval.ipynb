```python


import numpy as np # linear algebra
import pandas as pd # data processing, CSV file I/O (e.g. pd.read_csv)

df = pd.read_csv('../input/functional-preprocess-nowindowfeaturesextraction/trainProductUserdf_addedSomeFeatures_addedDummyPrediction.csv')
target = pd.read_csv('../input/functional-preprocess-nowindowfeaturesextraction/targetFor_trainProductUserdf.csv')
```

```python
df = df[['user_id','product_id','dummyPrediction_goneDay','dummyPrediction_datetime']]
df.head()
```

```python
target = target[['user_id','product_id','GoneDays']]
target.head()
```

```python
result = pd.merge(left=df,right=target,how='inner',on=['user_id','product_id'])
result.head()
```
